```julia
using ArraySignalProcessing
using LinearAlgebra

fc = 150e6

# Spacing:  1m is /2 for 150MHz 
# Oriented along yaxis to match Matlab behavior
pa = IsotropicArray([0 0 0 0;
                    -1.5 -0.5 0.5 1.5])
 = 1

# test signal generation
Rss = I(2)
 = deg2rad.([10, 65])
d = length()
N = 2000
SNR = 5
X = unconditional_signals(pa, Rss, N, SNR, , fc)

grid = Vector(-90:90)
A = steer(pa, deg2rad.(grid), fc)
spectrum = omp(X, A, d)
est = grid[spectrum .> 0]
display(est)

using Plots
import Measures

# clip OLS spectrum to show the zero-entries
spectrum_clipped = max.(spectrum, 1e-6)

p1 = plot(grid, 10*log10.((spectrum_clipped)),marker=:dot, legend=false)
plot!(title="OLS Spectrum", margin=5*Measures.mm)
xlabel!("Azimuth in degree")
ylabel!("Power in dB")
```

```julia
using CUDA
using ArraySignalProcessing
using LinearAlgebra

fc = 150e6

# Spacing:  1m is /2 for 150MHz 
# Oriented along yaxis to match Matlab behavior
pa = IsotropicArray(cu([0 0 0 0;
                    -1.5 -0.5 0.5 1.5]))
 = 1

# test signal generation
Rss = I(2)
 = cu(deg2rad.([10, 65]))
d = length()
N = 2000
SNR = 5
X = unconditional_signals(pa, Rss, N, SNR, , fc)

grid = cu(collect(Vector(-90:90)))
A = steer(pa, deg2rad.(grid), fc)
spectrum = omp(X, A, d)
est = grid[spectrum .> 0]
display(est)

using Plots
import Measures

# clip OLS spectrum to show the zero-entries
spectrum_clipped = max.(spectrum, 1e-6)

p1 = plot(Array(grid), Array(10*log10.(spectrum_clipped)),marker=:dot, legend=false)
plot!(title="OLS Spectrum", margin=5*Measures.mm)
xlabel!("Azimuth in degree")
ylabel!("Power in dB")
```
